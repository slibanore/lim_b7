```python
def f_esc(z,SFR_M):

    if z < 1.5:
        csi = 1.6
        z0 = 0.87
        zeta = 1/4.
        f0 = 0.35
        SFR0 = 1.29 # Msun yr-1
        t = 1.25
    else:
        csi = 1.6
        z0 = 3.125
        zeta = 1/4.
        f0 = 0.18
        SFR0 = 1.29 # Msun yr-1
        t = 0.875

    f = ((1 + np.exp(-csi*(z-z0)))**(-zeta)*(f0 + (1-f0)/(1+(SFR_M/SFR0)**t)))**2 # erg s-1

    return f

def ergsec_to_Lsun(L_ergsec):

    L_Lsun = L_ergsec / 3.826e33

    return L_Lsun

def Lsun_to_ergsec(L_Lsun):

    L_ergsec = L_Lsun * 3.826e33

    return L_ergsec
```
